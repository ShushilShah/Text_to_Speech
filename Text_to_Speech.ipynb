# Project: Conversion of Given text to Speech

# Problem Statement:
Convert any text data into speech format

# Project Workflow:
1. setup necessary libraries
2. import or provide text data
3. apply method
4. output

# 1. Set up necessary libraries
For the given problem stated, gTTS will be required libraries for this problem

```python
from gtts import gTTS
from IPython.display import Audio
```

# 2. Importing or  Providing text data

```python
tts = gTTS('Hello! my name is shushil shah a software engineer.')
```

# 3. Applying method

```python
tts.save('audio.wav')
sound_file = 'audio.wav'
```

```python
#wav is the extension for saving audio content
```

# 4. output

```python
Audio(sound_file,autoplay=True)
```
